```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('diabetes.csv')

# Scatter plot
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df['Age'], y=df['Glucose'])
plt.title('Scatter plot: Age vs Glucose')
plt.show()
```

```python

# Pair plot
sns.pairplot(df)
plt.show()
```

```python

# Histogram
df.hist(figsize=(10, 8))
plt.show()
```

```python

# Box plot
plt.figure(figsize=(6, 4))
sns.boxplot(data=df)
plt.title('Box plot of features')
plt.xticks(rotation=90)
plt.show()

# Shape of dataset
print("Shape of dataset:", df.shape)

# Filtering based on conditions
outcome_0 = df[df['Outcome'] == 0][['Glucose', 'BloodPressure']]
outcome_1 = df[df['Outcome'] == 1][['BMI', 'SkinThickness', 'Insulin']]

# Display filtered data
print("Outcome = 0: Glucose & BloodPressure")
print(outcome_0.head())

print("\nOutcome = 1: BMI, SkinThickness, Insulin")
print(outcome_1.head())
```

```python

# Condition: If Insulin == 0
insulin_zero = df[df['Insulin'] == 0]
print("\nRows where Insulin = 0:")
print(insulin_zero.head())

# Condition: If Glucose > 150
high_glucose = df[df['Glucose'] > 150][['Age', 'Outcome']]
print("\nRows where Glucose > 150:")
print(high_glucose.head())
```
