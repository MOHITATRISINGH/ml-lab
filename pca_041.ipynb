```python
import numpy as np
import pandas as pd

# Step 1: Load the dataset

df = pd.read_csv('Iris.csv')
print("Dataset loaded successfully.")

# Step 2: Select feature columns
features = df.drop(columns=["Id", "Species"])
feature_names = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"]
features.columns = feature_names
print("Selected feature columns: ", feature_names)

# Step 3: Standardizing the dataset
mean_vec = np.mean(features, axis=0)
std_vec = np.std(features, axis=0)
standardized_features = (features - mean_vec) / std_vec
print("Data standardized.")

# Step 4: Compute covariance matrix
cov_matrix = np.cov(standardized_features, rowvar=False)
print("Covariance matrix computed.")

# Step 5: Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print("Eigenvalues and eigenvectors computed.")

# Step 6: Sort eigenvectors by descending eigenvalues
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvectors_sorted = eigenvectors[:, sorted_indices]
print("Eigenvectors sorted by eigenvalues.")

# Step 7: Select the top 'num_components' eigenvectors
num_components = 2
selected_eigenvectors = eigenvectors_sorted[:, :num_components]
print(f"Selected top {num_components} principal components.")

# Step 8: Project data onto the new feature space
pca_result = np.dot(standardized_features, selected_eigenvectors)
print("Data projected onto new feature space.")

# Step 9: Convert result to DataFrame and display the first few rows
pca_df = pd.DataFrame(pca_result, columns=["PC1", "PC2"])
print("PCA transformation completed.")
pca_df
```
